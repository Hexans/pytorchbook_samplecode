# Chapter2 PyTorchの基本
## 1. テンソル（TENSOR）

```python
# 必要なパッケージをインポート
import torch
import numpy as np
```

## 1.1. Tensorの生成

```python
# listを渡してTensorを生成
x = torch.tensor([1, 2, 3])
print(x)
```

```python
# listを入れ子にしてTensorを生成
x = torch.tensor([[1, 2, 3], [4, 5, 6]])
print(x)
```

```python
x = torch.tensor([[1, 2, 3], [4, 5, 6]])
print(x.size())  # Tensorの形状を確認
```

```python
# データ型を指定せずにTensorを生成
x1 = torch.tensor([[1, 2, 3], [4, 5, 6]])

# dtypeを指定して64ビット浮動小数点数型のTensorを生成
x2 = torch.tensor([[1, 2, 3], [4, 5, 6]], dtype=torch.float64)

# torchのメソッドから64ビット浮動小数点数型のTensorを生成
x3 = torch.DoubleTensor([[1, 2, 3], [4, 5, 6]])

# dtypeの確認
print(x1.dtype)  # データ型指定なし
print(x2.dtype)  # データ型指定
print(x3.dtype)  # データ型指定
```

```python
# 0から9までの1次元Tensorを生成
x = torch.arange(0, 10)
print(x)
```

```python
# 0から始まって9まで2ずつ増えていく1次元Tensorを生成
x = torch.linspace(0, 9, 5)
print(x)
```

```python
# 0から1の間の乱数を生成
x = torch.rand(2, 3)
print(x)
```

```python
# 2x3の零テンソルを生成
x = torch.zeros(2, 3)
print(x)
```

```python
# 形状が2x3で要素がすべて1のテンソルを生成
x = torch.ones(2, 3)
print(x)
```

```python
x = torch.tensor([1, 2, 3]).to('cuda')
print(x.device)
```

## 1.2. Tensorとndarrayの変換

```python
# ndarrayの生成
array = np.array([[1,2,3],[4,5,6]])
print(array)
```

```python
# ndarrayからTensorへ変換
tensor = torch.from_numpy(array)
print(tensor)
```

```python
# Tensorからndarrayへ変換
tensor2array = tensor.numpy()
print(tensor2array)
```

## 1.3. Tensorの操作

```python
# インデックスの指定
x = torch.tensor([[1, 2, 3], [4, 5, 6]])
print(x[1, 2])
```

```python
# スライスで要素を取得
x = torch.tensor([[1, 2, 3], [4, 5, 6]])
print(x[1, :])
```

```python
# 2x3から3x2のTensorに変換
x = torch.tensor([[1, 2, 3], [4, 5, 6]])
x_reshape = x.view(3, 2)
print(x)  # 変換前の2x3のTensor
print(x_reshape)  # 変換後の3x2のTensor
```

## 1.4. Tensorの演算

```python
# Tensorとスカラーの四則演算
x = torch.tensor([[1, 2, 3], [4, 5, 6]], dtype=torch.float64)
print(x + 2)  # 足し算
print(x - 2)  # 引き算
print(x * 2)  # 掛け算
print(x / 2)  # 足し算
```

```python
# Tensor同士の四則演算
x = torch.tensor([[1, 2, 3], [4, 5, 6]], dtype=torch.float64)
y = torch.tensor([[4, 5, 6], [7, 8, 9]], dtype=torch.float64)
print(x + y)  # 足し算
print(x - y)  # 引き算
print(x * y)  # 掛け算
print(x / y)  # 割り算
```

```python
x = torch.tensor([[1, 2, 3], [4, 5, 6]], dtype=torch.float64)
print(torch.min(x))  # 最小値
print(torch.max(x))  # 最大値
print(torch.mean(x))  # 平均値
print(torch.sum(x))  # 合計値
```

```python
x = torch.tensor([[1, 2, 3], [4, 5, 6]], dtype=torch.float64)
print(torch.sum(x).item())  # 合計値
```
